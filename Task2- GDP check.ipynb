# Data Cleaning & Preparing

```python
# Checking for null values
df.isna().mean().round(4)*100
```

```python
df['West Bengal1']
```

```python
df.drop('West Bengal1',axis=1, inplace=True) # Dropping null column
```

```python
# Handle Missing values
df['Arunachal Pradesh'].fillna(df['Arunachal Pradesh'].mean(), inplace=True)
df['Assam'].fillna(df['Assam'].mean(), inplace=True)
df['Bihar'].fillna(df['Bihar'].mean(), inplace=True)
df['Goa'].fillna(df['Goa'].mean(), inplace=True)
df['Gujarat'].fillna(df['Gujarat'].mean(), inplace=True)
df['Himachal Pradesh'].fillna(df['Himachal Pradesh'].mean(), inplace=True)
df['Jammu & Kashmir'].fillna(df['Jammu & Kashmir'].mean(), inplace=True)
df['Jharkhand'].fillna(df['Jharkhand'].mean(), inplace=True)
df['Karnataka'].fillna(df['Karnataka'].mean(), inplace=True)
df['Kerala'].fillna(df['Kerala'].mean(), inplace=True)
df['Maharashtra'].fillna(df['Maharashtra'].mean(), inplace=True)
df['Manipur'].fillna(df['Manipur'].mean(), inplace=True)
df['Mizoram'].fillna(df['Mizoram'].mean(), inplace=True)
df['Nagaland'].fillna(df['Nagaland'].mean(), inplace=True)
df['Punjab'].fillna(df['Punjab'].mean(), inplace=True)
df['Rajasthan'].fillna(df['Rajasthan'].mean(), inplace=True)
df['Sikkim'].fillna(df['Sikkim'].mean(), inplace=True)
df['Tripura'].fillna(df['Tripura'].mean(), inplace=True)
df['Uttar Pradesh'].fillna(df['Uttar Pradesh'].mean(), inplace=True)
df['Uttarakhand'].fillna(df['Uttarakhand'].mean(), inplace=True)
df['Andaman & Nicobar Islands'].fillna(df['Andaman & Nicobar Islands'].mean(), inplace=True)
df['Chandigarh'].fillna(df['Chandigarh'].mean(), inplace=True)
```
